# Import geoJSON to do mapping

```python
import geopandas as gpd
```

```python
# upload the counties geojson file from LA times from my data folder
cty = gpd.read_file('../../data/counties-2012.geojson')
cty.head()
```

```python
# get info on the dataset
cty.info(verbose=True, null_counts=True)
```

good, 58 is the number of counties that I know from using the other dataset. All fields are filled too

```python
# get rid of everything except FIPS and geometry. FIPS in this dataset is the external_id, but in the pop dataset it is called Geo_FIPS
cty=cty[['external_id','geometry']]
cty.head()
```

```python
# need to rename external_id to match pop dataset
cty.columns=['Geo_FIPS','geometry']
```

```python
cty.head()
```

```python
# what does it look like visually
cty.plot(figsize=(12,10))
```

```python
# need to merge pop and cty by the Geo_FIPS field
cty_pop = cty.merge(pop,on="Geo_FIPS")
```

```python
# make sure it joined correctly
cty_pop.head()
```

```python
# chloropleth map of the counties' population density 
cty_pop.plot(figsize=(12,10), 
             column='Pop_Den_sqmi',
             legend=True,
             scheme='NaturalBreaks')
```

```python
cty_pop.plot(figsize=(12,10), 
             column='Pop_Den_sqmi',
             legend=True,
             scheme='equal_interval')
```

```python
cty_pop.plot(figsize=(12,10), 
             column='Pop_Den_sqmi',
             legend=True,
             scheme='quantiles')
```

wow, the three types of schemes are drastically different. In this case equal interval is basically useless.

```python
# now I'll look at total population
cty_pop.plot(figsize=(12,10), 
             column='Total_Pop',
             legend=True,
             scheme='NaturalBreaks')
```

```python
cty_pop.plot(figsize=(12,10), 
             column='Total_Pop',
             legend=True,
             scheme='equal_interval')
```

```python
cty_pop.plot(figsize=(12,10), 
             column='Total_Pop',
             legend=True,
             scheme='quantiles')
```

Again, equal interval is kind of useless. However, natural breaks and quantiles show very different stories. Natural breaks clearly shows LA County is the most populous, and that SoCal and the Bay Area are basically the only higer-populated areas. But the quantiles version shows the whole southland, the bay, the Central Coast, and the Central Valley are pretty populated.
